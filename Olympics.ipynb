```python
df['Gender'].value_counts().plot(kind='pie', fontsize=14, autopct='%3.1f%%', figsize=(12,7), shadow=True, startangle=135, legend=True, cmap='Blues_r')
plt.ylabel('Gender')
plt.title('Number of Men and Women')
```

```python
medals = np.array(df['Medal'])
medals = np.unique(medals)
counts = {}
for i in medals:
  medal_data = df[df['Medal']==i]
  index = medal_data.index
  rows = len(index)
  counts.update({i:rows})
print(dict(sorted(counts.items(), key = lambda x: x[1], reverse = True)))
```

```python
sports = np.array(df['Sport'])
sports = np.unique(sports)
counts = {}
for i in sports:
  sport_data = df[df['Sport']==i]
  index = sport_data.index
  rows = len(index)
  counts.update({i:rows})
print(dict(sorted(counts.items(), key = lambda x: x[1], reverse = True)))
```

```python
df.Sport.value_counts().sort_values(ascending = False).plot(kind ='bar', figsize=(20,8), fontsize=12, color='green')
plt.title("Fequencies of Sports", fontsize=18,fontweight='bold')
plt.ylabel("Sports Count", size= 20)
plt.xlabel("Sports", size = 20)
```

```python
disciplines = np.array(df['Discipline'])
disciplines = np.unique(disciplines)
counts = {}
for i in disciplines:
  discipline_data = df[df['Discipline']==i]
  index = discipline_data.index
  rows = len(index)
  counts.update({i:rows})
print(dict(sorted(counts.items(), key = lambda x: x[1], reverse = True)))
```

```python
df.Discipline.value_counts().sort_values(ascending = False).plot(kind ='bar', figsize=(20,10), fontsize=12, color='blue')
plt.title("Fequencies of Disciplines", fontsize=18,fontweight='bold')
plt.ylabel("Count", size= 20)
plt.xlabel("Disciplines", size = 20)
```
